## Categorical Feature Treatment

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data_tre = cre_data.copy()
```

```python
data_tre.drop('loan_grade', axis = 1 , inplace= True)
```

```python
person_home_ownership = pd.get_dummies(data_tre['person_home_ownership'], drop_first= True).astype(int)
loan_intent = pd.get_dummies(data_tre['loan_intent'], drop_first= True).astype(int)
data_tre['cb_person_default_on_file_binary'] = np.where(data_tre['cb_person_default_on_file'] == 'Y',1,0)
data_tre.head()
```

```python
person_home_ownership.head()
```

```python
data_scale = data_tre.drop(['person_home_ownership','loan_intent','cb_person_default_on_file','cb_person_default_on_file_binary'], axis = 1)
```

```python
data_scale.head()
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(data_scale)
scaled_data = pd.DataFrame(scaled , columns= ['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_status', 'loan_percent_income',
       'cb_person_cred_hist_length'])
scaled_data.head()
```

```python
scaled_data.shape
```

```python
data_combined = pd.concat([scaled_data,person_home_ownership,loan_intent], axis = 1)
data_combined['cb_person_default_on_file'] = data_tre['cb_person_default_on_file_binary']
data_combined['loan_status'] = data_tre['loan_status']
data_combined
```

```python
data_combined['loan_status'].value_counts()
```

```python
data_combined = data_combined.reset_index(drop = True)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score , r2_score , accuracy_score , precision_score ,roc_auc_score , mean_squared_error,mean_absolute_error
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
```

```python
X_train ,X_test, y_train ,y_test = train_test_split(X , y , random_state= 42 , test_size= 0.2)
```

```python
lg = LogisticRegression()
lg.fit(X_train,y_train)
```

```python
predict_lg = lg.predict(X_test)
predict_lg
```

```python
lg.score(X_train,y_train)
```

```python
print(classification_report(y_test , predict_lg))
```

```python
print(lg.coef_[0])
```

```python
feature_importance_lg =  pd.DataFrame({
    'features': X.columns,
    'Score': lg.coef_[0]
})
```
